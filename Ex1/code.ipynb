```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

csv_path = 'customer_data.csv'
df = pd.read_csv(csv_path)
```

```python
df['fea_2'] = df['fea_2'].fillna(df['fea_2'].mean())
labels_df = df['label']
labels = labels_df.to_numpy()

x_df = df.iloc[:,1:]
x = x_df.to_numpy()
```

```python
x_train, x_test, labels_train, labels_test = train_test_split(x, labels, test_size=0.2, random_state=42)
```

```python
#Data scaling
mean_cols = np.nanmean(x_train, axis=0)
std_cols = np.nanstd(x_train, axis=0)

x = (x - mean_cols)/std_cols
```

```python
num = 0
K = 5

for i in range(len(labels_test)):
    distances = np.sqrt(np.sum((x_train - x_test[i])**2, axis=1))
    
    k_nearest_indices = np.argsort(distances)[:K]
    k_nearest_labels = labels_train[k_nearest_indices]
    
    prediction = np.bincount(k_nearest_labels).argmax()
    if prediction == labels_test[i]:
        num += 1
    
print(f"Accuracy: {num/len(labels_test):.4f}")
```
